## Введение

Большая совокупность случайных явлений или величин подчиняется стетистическим закономерностям.

Пусть имеется N одинаковых систем, для которых определённая величина х может принимать дискретные значения х_1, х_2 и т.д.

Проведём измерения этой величины в каждой системе. N_i измерений дали результат x_i 

$$ \frac {N_i}{N}$$

Относительная частота появления результата x_i

$$ lim_{N->\infty}\frac{N-i}{N}=P_i$$ - вероятность появления результата x_i

Вероятность попадания величины х в интервале от х до х+dх зависит от характера величины и от ширины выбранного интервала:

$$\frac{dN}{N} = f(x)dx$$

dN - число измерений, в которых величина х имела значение в интервале от х до dх.

$$ f(x) = \frac{dN}{N}*\frac{1}{dx}$$

функция распределения величины х или плотность вероятности.

Среднее значение любой величины L(x), если известна функция распределения f(x) имеет вид:

$$ <L> = \int_0^\inftyl(x)f(x)dx$$
